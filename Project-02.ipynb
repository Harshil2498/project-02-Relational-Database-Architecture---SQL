```python
import sqlite3
import pandas as pd

df = pd.read_csv('transaction.csv')

conn = sqlite3.connect('finance_health.db')

df.to_sql('finances', conn, if_exists='replace', index=False)

print("Database created and data uploaded successfully!")
```

```python
query = '''
SELECT 
    user_id, 
    monthly_income, 
    monthly_expense_total, 
    financial_stress_level
FROM finances
WHERE monthly_expense_total > monthly_income
AND fraud_flag = 1;
'''

# Execute and view results
high_risk_df = pd.read_sql_query(query, conn)
print(high_risk_df.head())
```

```python
query_ranking = '''
SELECT user_id, income_type, savings_rate, credit_score,
       RANK() OVER (PARTITION BY income_type ORDER BY savings_rate DESC) as rank_within_group
FROM finances
WHERE credit_score > 700
LIMIT 15;
'''
ranking_df = pd.read_sql_query(query_ranking, conn)
print(ranking_df)
```

```python
query_risk = '''
SELECT user_id, monthly_income, monthly_expense_total
FROM finances
WHERE monthly_expense_total > monthly_income
AND fraud_flag = 1;
'''
risk_df = pd.read_sql_query(query_risk, conn)
print(risk_df)
```
